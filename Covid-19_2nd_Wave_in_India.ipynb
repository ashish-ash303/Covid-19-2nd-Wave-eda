## Finding Insights

### Which State/UTs has highest recovery rate ?

```python
recovery_df=covid_df.sort_values('Discharge Ratio (%)',ascending=False)
recovery_df.head()
```

> Dadra and Nagar Haveli and Daman and Diu has highest recovery rate

## Which State/UTs has most active Cases ?

```python
active_df=covid_df.sort_values('Active',ascending=False)
active_df.head()
```

> Kerela has most  active Cases. The reasons can be its population as people from world or country visits this state in search  of job or for some vaccation trip 


## Which State/UTs has minimum Death Ratio(%)

```python
deathratio_df=covid_df.sort_values('Death Ratio (%)',ascending=True)
deathratio_df.head()
```



Dadra and Nagar Haveli and Daman and Diu has minimum Death ratio and highest Discharge ratio also have only 4 active cases.
The most probable reasons are:


    1.Daman and Diu  has some good medical infrastructure.
    
    2.Less population. 
    
    3.People are more aware about there Responsiblity and follows all the norms,maintiains social distancing, Waer masks..
    
    4.Taking Vaccinations.
    
    
 
These are some reason why active cases and Death ratio in Dadra and Nagar Haveli and Daman and Diu are less While discharge ratio are high.
    

## Which State has highest Discharged Patients??

```python
discharged_df=covid_df.sort_values('Discharged',ascending=False)
discharged_df.head()
```

## Which State has highest Total Cases ??

```python
totalcases_df=covid_df.sort_values('Total Cases',ascending=False)
totalcases_df.head()
```

Maharashtra State has highest total cases while also have highest discharged patient. The most Probable reason is it's Population

## Which State/UTs has highest Active Ratio ??

```python
activeratio_df=covid_df.sort_values('Active Ratio (%)',ascending=False)
activeratio_df.head()
```

```python
death_df=covid_df.sort_values("Deaths",ascending=True)
death_df.head()
```

Mizoram has highest Active case ratio.
